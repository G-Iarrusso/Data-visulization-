Introduction and Motivation

The research hypothesis is that because Ontario has major cities such as Toronto and Hamilton with high populations such that Ontario should have one of, if not, the highest crime severity index of all of Canada. Furthermore, once the yearly indexes for Ontario have been compared to other provinces and territories, the major cities in Ontario will be analyzed to see what their crime severity index is, compared to other cities within Ontario. There was interest in doing this because the constant stories heard from the news about violent crimes in Ontario and this visualization will compare how much it differs from other places in Canada. Note that crime severities will be compared from the years 2008 - 2012 inclusive. Furthermore, examining if there is a trend in crime severities per year both nationally and in Ontario, this will be cross examined with a overall 5 year average.

```python
import pandas as pd
from matplotlib import pyplot
import numpy as np
df_city = pd.read_csv('ontario_data.csv')
df_pro = pd.read_csv('provincial_data.csv')
```

Structure

ontario_data.csv is the file containing the crime severity indexes of 15 major Ontario cities
provincial_data.csv is the file containing all the crime severities for the provinces and territories

Crime severity is an index that measures how bad a crime was, the major metric for this is the time sentenced and the time served. The longer and more serious the crime the higher the value. The index starts at 100 when first implemented in 2006. 

Note that for all bar graphs, the order of cities/provinces/territories is based on the 5 year average with ascending order.

```python
pyplot.rcParams.update({'font.size': 22})
ax = df_pro.plot.barh(x = 'Location',y = ['2012','2011','2010','2009','2008'], title ="Crime Severity by Province", width = .8,figsize = (20,20),edgecolor = "white");
ax.set_xlabel("Severity Index")
ax.set_ylabel("Provinces/Territories")

ay = df_pro.plot.box(x = 'Location',figsize = (15,15));
ay.set_xlabel("Year")
ay.set_ylabel("Severity Index")
ay.set_title("National Crime Severity by Year")
```

After examination of the grouped bar graph Ontario, the crime severity index has not only been steadily decreasing each year, but it is also the lowest of all provinces and territories. Other findings from this graph show: the highest three locations are the territories with Nunavut and Northwest territories appearing to be outliers with crime severities much higher than any other areas. Another observation is that for most areas, there is an overall decrease in crime severity between 2008's and 2012's indexes. This leads to the boxplot where it is clear that overall Canada’s crime severity  over 2011 and 2012 falls below the average for our time span, suggesting that overall crime severity has been steadily decreasing. 

```python
pyplot.rcParams.update({'font.size': 22})
az = df_city.plot.barh(x = 'City',y = ['2012','2011','2010','2009','2008'], title ="Crime Severity by City", width = 0.8 ,figsize = (20,20),edgecolor = "White");
az.set_xlabel("Severity Index")
az.set_ylabel("Ontario Cities")

aa = df_city.plot.box(x = 'City',figsize = (15,15));
aa.set_xlabel("Year")
aa.set_ylabel("Severity Index")
aa.set_title("Ontario Crime Severity by Year")
```

The top five cities in Ontario in 2011 by population is in descending order: Toronto, Ottawa-Gatineau (On/Qu), Hamilton, Kitchener and London. From examination of the grouped bar graph, there is no correlation with population having higher or lower crime severities. Toronto and Ottawa-Gatineau (both Ontario and Quebec sides) are on the lower end of provincial severity. Kitchener is the exact middle of crime severity in the province for the listed cities. Hamilton and London are in the top half of crime severity in Ontario. Furthermore, one can examine the boxplot and find that there is a slight increase from 2008 to 2009, and then a greater decrease from 2009 to 2012. Similar to the country's severity indexes from 2011 and 2012 Ontario’s indexes also fall below the provinces average index over the 5 year period.

Conclusion
